95% konfidensinterval for democrats

```python
import scipy.stats as stats
import math
import numpy as np
import statsmodels.stats.proportion as smprop
```

Republican

```python
# Data 
x = 27  # Antal succeser 
n = 60  # Antal observationer 


phat = x / n

#konfidensinterval
se = np.sqrt(phat * (1 - phat) / n)  
ci_lower = phat - 1.96 * se
ci_upper = phat + 1.96 * se


print(f"Observeret proportion: {phat:.4f}")

print(f"95% Konfidensinterval: ({ci_lower:.4f}, {ci_upper:.4f})")
```

Democrats

```python
# Data
x = 33  # Antal succeser 
n = 60  # Antal observationer 


phat = x / n

# Konfidensinterval 
se = np.sqrt(phat * (1 - phat) / n)  # Standardfejl
ci_lower = phat - 1.96 * se
ci_upper = phat + 1.96 * se


print(f"Observeret proportion: {phat:.4f}")

print(f"95% Konfidensinterval: ({ci_lower:.4f}, {ci_upper:.4f})")
```

Null-hypothesis test

```python
z_obs,p_value = smprop.proportions_ztest(count = [27, 33], nobs = [60], value=0, prop_var=0)
print(z_obs, p_value)
```

```python
#plots created with assistance from chat GPT
import matplotlib.pyplot as plt

# Republican policy topics
republican_data = {
    "Strong National Defense": 2,
    "Tax Cuts & Deregulation": 1,
    "Immigration Enforcement & Border Security": 7,
    "Religious Liberty & Family Values": 1,
    "Second Amendment Rights": 0,
    "Healthcare Reform & Free Market Approach": 0,
    "Pro-Life Stance": 1,
    "Judicial Appointments & Originalism": 1,
    "Education & School Choice": 2,
    "Energy Independence": 3,
    "Law & Order": 1,
    "Economic Growth & Job Creation": 2,
    "Budget & Government Spending": 0,
    "Trade Policy & American Manufacturing": 0,
    "Foreign Policy & National Sovereignty": 6,
    "Non political": 33
}

# Democratic policy topics
democratic_data = {
    "Healthcare Expansion": 2,
    "Climate Action": 1,
    "Economic Equity & Worker Protections": 8,
    "Voting Rights & Electoral Reform": 4,
    "Racial Justice & Equity": 1,
    "Immigration Reform": 0,
    "Gun Violence Prevention": 0,
    "Women’s Rights & Reproductive Health": 3,
    "Education & Student Debt Relief": 0,
    "LGBTQ+ Equality": 0,
    "Criminal Justice & Policing Reform": 5,
    "Infrastructure & Green Jobs": 4,
    "Tax Reform": 1,
    "Health & Social Safety Nets": 4,
    "Foreign Policy & Diplomacy": 0,
    "Non political": 27
}

def create_bar_chart(data_dict, title, bar_color='skyblue'):
   
    categories = list(data_dict.keys())
    values = list(data_dict.values())
    
    
    plt.figure(figsize=(10, 5))  
    bars = plt.bar(categories, values, color=bar_color, edgecolor='black')
    
    
    for bar in bars:
        height = bar.get_height()
        plt.text(
            bar.get_x() + bar.get_width() / 2, 
            height + 0.5,  
            str(height), 
            ha='center', 
            va='bottom'
        )
    
    
    plt.title(title, fontsize=14, fontweight='bold')
    plt.xlabel('Policy Categories')
    plt.ylabel('Number of Topics')
    
    
    plt.xticks(rotation=45, ha='right')
    
   
    plt.tight_layout()
    plt.show()


create_bar_chart(republican_data, "Republican Policy Topics", bar_color='red')

create_bar_chart(democratic_data, "Democratic Policy Topics", bar_color='darkblue')
```
